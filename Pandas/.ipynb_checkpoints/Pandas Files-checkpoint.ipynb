# HeatMaps

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
oo=pd.read_csv('/home/shubham/Downloads/Ex_Files_Pandas_EssT/ExerciseFiles/data/olympics.csv')
```

```python
a=oo[oo.Edition==2008]
```

```python
b=a.groupby(['NOC','Medal']).size()
b.head()
```

```python
c=b.unstack('Medal',fill_value=0)
c.head()
```

```python
d=c.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
d.head()
```

```python
sns.heatmap(d)
```

```python
e=d.transpose()
e
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(d)
```
